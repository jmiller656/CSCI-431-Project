```python
import keras.backend as K
from keras.layers.core import Lambda
from keras.layers.normalization import BatchNormalization

from nalu import NALU
model = Sequential()
model.add(InputLayer(input_shape=(None, None, 1)))
model.add(Conv2D(2**3, (3, 3), activation='relu', padding='same', strides=1))
model.add(Conv2D(2**4, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
model.add(Conv2D(2**4, (3, 3), activation='relu', padding='same', strides=1))
model.add(Conv2D(2**5, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
model.add(Conv2D(2**5, (3, 3), activation='relu', padding='same', strides=1))
model.add(Conv2D(2**6, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
model.add(Conv2D(2**6, (3, 3), activation='relu', padding='same', strides=1))
#model.add(BatchNormalization())
#model.add(UpSampling2D((2, 2)))
model.add(Conv2D(2**6, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
#model.add(UpSampling2D((2, 2)))
model.add(Conv2D(2**5, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
#model.add(UpSampling2D((2, 2)))
model.add(Conv2D(2**5, (3, 3), activation='relu', padding='same'))
#model.add(BatchNormalization())
#model.add(UpSampling2D((2, 2)))
model.add(Conv2D(3, (3,3), activation=None, padding='same'))
model.add(Lambda(lambda x: K.clip(x, 0, 1)))
```

```python
# Finish model
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(x=x, y=y, validation_data=(test_x,test_y), batch_size=32, epochs=10, shuffle=True)
```
